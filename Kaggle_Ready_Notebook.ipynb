### 3.2 Column Descriptions (All 46 Columns)

| # | Column | Type | Description |
|---|---|---|---|
| 1 | `ID` | str | Unique alphanumeric record identifier (e.g., A-3456789). Primary key — no analytical value on its own. |
| 2 | `Source` | str | API/provider that reported the accident (MapQuest, Bing, or both). Useful for studying data collection bias. |
| 3 | `Severity` | int | Traffic impact score from 1 (minor delay) to 4 (road closure). Primary ML target — heavily skewed toward 2. |
| 4 | `Start_Time` | datetime | Exact timestamp when the accident was first detected. Rich source for time-based features. |
| 5 | `End_Time` | datetime | Timestamp when traffic flow returned to normal. Used to calculate accident duration. |
| 6 | `Start_Lat` | float | Latitude of the accident start point (24°–50° for contiguous US). Core geospatial column. |
| 7 | `Start_Lng` | float | Longitude of the accident start point (-125° to -65°). Negative values since US is in western hemisphere. |
| 8 | `End_Lat` | float | Latitude of the far end of the affected road segment. Often missing or equal to Start_Lat for minor accidents. |
| 9 | `End_Lng` | float | Longitude of the far end of the affected road segment. Works with End_Lat to define the impacted road segment. |
| 10 | `Distance(mi)` | float | Length of road (miles) affected by the accident. Proxy for severity — highways tend to have larger values. |
| 11 | `Description` | str | Auto-generated natural language summary from the API. Quality varies by source. Useful for NLP tasks. |
| 12 | `Street` | str | Street or highway name (e.g., I-95 N, Main St). Useful for identifying accident-prone roads. |
| 13 | `City` | str | City where the accident occurred. Miami, Houston, LA, Charlotte are the most frequent. |
| 14 | `County` | str | County of the accident. Useful for regional analysis between city and state granularity. |
| 15 | `State` | str | Two-letter US state abbreviation (e.g., CA, TX). One of the cleanest and most complete columns. |
| 16 | `Zipcode` | str | Postal ZIP code (5-digit or ZIP+4 format). Useful for hyper-local analysis. |
| 17 | `Country` | str | Always `US`. Zero variance — no analytical value. |
| 18 | `Timezone` | str | Timezone of accident location (US/Eastern, US/Central, US/Mountain, US/Pacific). Critical for time analysis. |
| 19 | `Airport_Code` | str | IATA code of nearest weather station (not nearest airport). Links each record to its weather readings. |
| 20 | `Weather_Timestamp` | datetime | Timestamp of the weather observation. May differ from Start_Time — check the gap before using. |
| 21 | `Temperature(F)` | float | Air temperature in Fahrenheit. Ranges ~-30°F to 120°F. Strong correlation with road conditions. |
| 22 | `Wind_Chill(F)` | float | Perceived temperature accounting for wind speed. Only meaningful below 50°F. High null rate in warm records. |
| 23 | `Humidity(%)` | float | Relative humidity (0–100%). High values often correlate with fog or rain. |
| 24 | `Pressure(in)` | float | Atmospheric pressure in inches of mercury (~29.92 inHg normal). Rapid drops indicate incoming storms. |
| 25 | `Visibility(mi)` | float | Horizontal visibility in miles (0 = whiteout, 10 = clear). Below 1 mile strongly linked to severe accidents. |
| 26 | `Wind_Direction` | str | Compass direction wind blows from (N, NE, E, … Calm, Variable). Abbreviated — needs expansion. |
| 27 | `Wind_Speed(mph)` | float | Wind speed in mph. Above 30 mph affects vehicle control. Contains outliers above 300 mph (data errors). |
| 28 | `Precipitation(in)` | float | Precipitation amount in inches. Nulls and zeros often mean the same thing (no precipitation). |
| 29 | `Weather_Condition` | str | Weather label at accident time (Clear, Light Rain, Heavy Snow, Fog, etc.). 100+ unique values — needs grouping. |
| 30 | `Amenity` | bool | POI flag: amenity (park, shop, restaurant) nearby. 76L False vs 1L True — highly imbalanced. |
| 31 | `Bump` | bool | POI flag: speed bump or hump nearby. Only ~3,500 True values out of 7.7M. |
| 32 | `Crossing` | bool | POI flag: pedestrian or road crossing nearby. ~8 lakh True values. |
| 33 | `Give_Way` | bool | POI flag: yield/give-way sign nearby. Only 36,582 True values — highly imbalanced. |
| 34 | `Junction` | bool | POI flag: road junction nearby. 571,342 True — one of the stronger road feature predictors. |
| 35 | `No_Exit` | bool | POI flag: no-exit/dead-end road nearby. Only 19,545 True values. |
| 36 | `Railway` | bool | POI flag: railway level crossing nearby. Only 66,979 True values. |
| 37 | `Roundabout` | bool | POI flag: roundabout nearby. Only 249 True values — most extreme imbalance in dataset. |
| 38 | `Station` | bool | POI flag: transit/bus station nearby. 201,901 True — high pedestrian and congestion zone indicator. |
| 39 | `Stop` | bool | POI flag: stop sign nearby. Useful for identifying stop-sign intersection accidents. |
| 40 | `Traffic_Calming` | bool | POI flag: traffic calming infrastructure nearby (chicanes, raised crosswalks, etc.). Only 7,598 True. |
| 41 | `Traffic_Signal` | bool | POI flag: traffic signal (traffic light) nearby. 1,143,772 True — most analytically important POI column. |
| 42 | `Turning_Loop` | bool | POI flag: turning loop nearby. All False in this dataset — zero variance, drop this column. |
| 43 | `Sunrise_Sunset` | str | Day/Night based on official sunrise/sunset times for that location and date. Clean and complete. |
| 44 | `Civil_Twilight` | str | Day/Night based on civil twilight (sun 0°–6° below horizon). Captures dawn/dusk transition. |
| 45 | `Nautical_Twilight` | str | Day/Night based on nautical twilight (sun 6°–12° below horizon). Darker than civil twilight. |
| 46 | `Astronomical_Twilight` | str | Day/Night based on astronomical twilight (sun 12°–18° below horizon). Closest to full night classification. |

### 6.1 Time-Based Features

```python
## Accident time
df3['Accident_Duration(min)'] = round((df3['End_Time'] - df3['Start_Time']).dt.total_seconds()/60)
```

```python
## hour  
df3['hour'] = df3['Start_Time'].dt.hour
```

```python
## day of week
df3['Day_Of_Week'] = df3['Start_Time'].dt.day_name()
```

```python
## month
df3['Month_Name'] = df3['Start_Time'].dt.month_name()
```

```python
## Year
df3['Year'] = df3['Start_Time'].dt.year
```

```python
df3['Month_Number']=df3['Start_Time'].dt.month
```

```python
## Season
season = {
    12:'Winter', 1:'Winter', 2:'Winter',
    3:'Spring',  4:'Spring', 5:'Spring',
    6:'Summer',  7:'Summer', 8:'Summer',
    9:'Fall',   10:'Fall',  11:'Fall'
}
```

```python
df3['Season'] = df3['Month_Number'].map(season)
```

```python
df3.drop(columns='Month_Number',inplace=True)
```

```python
## is weekend
df3['is_weekend'] = df3['Start_Time'].dt.dayofweek >= 5
```

```python
## Time of the day
df3['time_of_day'] =pd.cut(df3['hour'],
    bins=[0,6,12,17,21,24],
    labels=['Night','Morning','Afternoon','Evening','Late Night'])
```

```python
## is rush hour
df3['is_Rush_Hour'] = df3['hour'].between(7,9) | df3['hour'].between(16,19)
```

### 6.2 Severity Label Mapping
> Mapped numeric severity (1–4) to descriptive labels: `Low`, `Medium`, `High`, `Critical`.

```python
## severity
df3['Severity'] = df3['Severity'].map(mapping)
```

### 6.3 Weather Risk Flags
> Binary flags to capture dangerous weather conditions clearly.


```python
## Poor Visisbility
df3['Poor_Visibility'] = df3['Visibility(mi)'] < 1
```

```python
## Is precipiting
df3['Is_Precipitaing'] = df3['Precipitation(in)'] > 0
```

```python
## temprature
df3['Temp_Category'] = pd.cut(df['Temperature(F)'] , bins = [-50 , 32,60,85,150] , labels=['Freezing' , 'Cold' , 'Mild' , 'Hot'])
```

### 6.4 POI Aggregation
> All 12 POI boolean columns (heavily imbalanced individually) are combined into two meaningful features:
> - `Nearby_POI_Count` — total number of POIs near the accident
> - `Has_Any_Poi` — simple binary flag


```python
## Nearby POI Count
cols = ['Amenity',
       'Bump', 'Crossing', 'Give_Way', 'Junction', 'No_Exit', 'Railway',
       'Roundabout', 'Station', 'Stop', 'Traffic_Calming', 'Traffic_Signal']
```

```python
df3['Nearby_POI_Count'] = df[cols].sum(axis=1)
```

```python
## has any poi
df3['Has_Any_Poi'] = df3['Nearby_POI_Count'] > 0
```
