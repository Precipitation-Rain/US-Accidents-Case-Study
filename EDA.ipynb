```python
def give_Summary(col_name):
    print("Statastical Summary\n")
    print(df[col_name].describe())
    print('_'*100)

    print("Distribution")
    df[col_name].plot(kind = 'hist')
    plt.show()
    print('_'*100)
    
    print("Outliers")
    (df[col_name].plot(kind = 'box'))
    plt.show()
    print('_'*100)

    print("Skewness")
    print(df[col_name].skew())
```

# Categoeical Column

```python
def get_summary_statestical(col_name):


    if(df[col_name].value_counts().count() > 10):

        print(df[col_name].value_counts().head(10))
        print('_' * 100)

        print("Value Counts")
        df[col_name].value_counts().head(10).plot(kind = 'bar')
        plt.show()
        print('_' * 100)

        print("Percentage Share")
        df[col_name].value_counts().head(10).plot(kind = 'pie' , autopct = '%.2f%%')
        plt.show()
        print('_' * 100)

        print("Top 10 categgories Sum",df[col_name].value_counts().head(10).sum())
        
    else:
        print(df[col_name].value_counts())
        print('_' * 100)


        df[col_name].value_counts().plot(kind = 'bar')
        plt.show()
        print('_' * 100)

        df[col_name].value_counts().plot(kind = 'pie' , autopct = '%.2f%%')
        plt.show()
        print('_' * 100)
```

```python
get_summary_statestical('Source')
```

## Bivariate Analysis

### Categorical to Numerical

```python
df = pd.read_csv("US_Accidents.csv", low_memory=False)

ct = pd.crosstab(df["Weather_Condition"], df["Severity"])
ct_pct = ct.div(ct.sum(axis=1), axis=0) * 100

# Keep top 10 weather conditions by volume
top10 = ct.sum(axis=1).nlargest(10).index
ct_pct = ct_pct.loc[top10]

plt.figure(figsize=(8, 6))
sns.heatmap(ct_pct, annot=True, fmt=".1f", cmap="YlOrRd",
            linewidths=0.4, cbar_kws={"label": "% within weather type"})
plt.title("Weather Condition × Accident Severity")
plt.xlabel("Severity")
plt.ylabel("Weather Condition")
plt.tight_layout()
plt.savefig("heatmap.png", dpi=150)
plt.show()
```
